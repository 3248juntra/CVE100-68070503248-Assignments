In this set of practice exercises, we will be working with a demographic data regarding the passengers aboard the Titanic.  Read in the data frame and use the head() method to check that it was read in correctly.

```python
# ...existing code...
import pandas as pd
import glob, os

# try to find a Titanic csv in the current folder (fallback to the first csv)
csvs = glob.glob(os.path.join(os.getcwd(), "*.csv"))
tfile = next((f for f in csvs if 'titanic' in os.path.basename(f).lower()), (csvs[0] if csvs else None))

if tfile is None:
    raise FileNotFoundError("No CSV file found in the working directory.")
df_titanic = pd.read_csv(tfile)
df_titanic.head()
# ...existing code...
```

Use the rename method to change the column "Name" to "Passenger_Name" and the column "Ticket" to "Ticket_Num".

```python
# ...existing code...
df_titanic.rename(columns={'Name': 'Passenger_Name', 'Ticket': 'Ticket_Num'}, inplace=True)
df_titanic.head()
# ...existing code...
```

Change the index of the df_titanic to the column PasengerID.

```python
# ...existing code...
#Write your code here
possible = ['PassengerId','PassengerID','Passenger_Id','Passenger Id','PasengerID','PasengerId']
col = next((c for c in possible if c in df_titanic.columns), None)
if col is None:
    col = next((c for c in df_titanic.columns if 'passeng' in c.lower()), None)
if col is None:
    raise KeyError("Passenger ID column not found in df_titanic.")
df_titanic.set_index(col, inplace=True)
df_titanic.head()
# ...existing code...
```

Select the name of passenger 896

```python
# ...existing code...
#Write your code here
# robustly find the name column
name_col = 'Passenger_Name' if 'Passenger_Name' in df_titanic.columns else next((c for c in df_titanic.columns if 'name' in c.lower()), None)

# get the original passenger id column name (it was set as the index)
pid_col = df_titanic.index.name
tmp = df_titanic.reset_index()

# try numeric 896 first, then string '896' as fallback
if pid_col in tmp.columns:
    row = tmp[tmp[pid_col] == 896]
    if row.empty:
        row = tmp[tmp[pid_col] == '896']
else:
    row = tmp[tmp.index == 896]
    if row.empty:
        row = tmp[tmp.index == '896']

name_896 = row[name_col].iloc[0] if (not row.empty and name_col in row.columns) else None
name_896
# ...existing code...
```

Change the index back to row numbers but making sure that we don't lose the PassengerId column. 

```python
# ...existing code...
#Write your code here
# Reset index to default row numbers and make sure the PassengerId column is kept
df_titanic.reset_index(inplace=True)

# Optional: standardize the passenger id column name to "PassengerId" if a candidate exists
pid_candidates = ['PassengerId','PassengerID','Passenger_Id','Passenger Id','PasengerID','PasengerId']
for c in pid_candidates:
    if c in df_titanic.columns and c != 'PassengerId':
        df_titanic.rename(columns={c: 'PassengerId'}, inplace=True)
        break

df_titanic.head()
# ...existing code...
```

How many missing entries are there in the Age column? Use the isnull() method and recall than Booleans are exactly like 1s (True) and 0s (False).

```python
# ...existing code...
#Write you code here
age_col = next((c for c in df_titanic.columns if 'age' in c.lower()), None)
if age_col is None:
    missing_age = 0
else:
    missing_age = df_titanic[age_col].isnull().sum()
missing_age
# ...existing code...
```

Compute the avg age of passengers ignoring the missing data.

```python
# ...existing code...
#Write your code here
age_col = next((c for c in df_titanic.columns if 'age' in c.lower()), None)
if age_col is None:
    avg_age = None
else:
    avg_age = df_titanic[age_col].mean()  # mean() ignores NaNs by default
avg_age
# ...existing code...
```

Using the fillna() method replace the missing values in the Age column with the mean.

```python
# ...existing code...
age_col = next((c for c in df_titanic.columns if 'age' in c.lower()), None)
if age_col is None:
    print("Age column not found")
else:
    mean_age = df_titanic[age_col].mean()
    df_titanic[age_col].fillna(mean_age, inplace=True)
df_titanic.head()
# ...existing code...
```

What is the average age of the 5 oldest passengers? The reset_index method will be helpful here.

```python
#Write your code here
age_col = next((c for c in df_titanic.columns if 'age' in c.lower()), None)
if age_col is None:
    print("Age column not found")
else:
    # Sort by age descending and get mean of top 5
    avg_oldest_5 = df_titanic[age_col].nlargest(5).mean()
    print(f"Average age of 5 oldest passengers: {avg_oldest_5:.2f}")
```
